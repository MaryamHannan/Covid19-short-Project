```python
import pandas as pd
```

```python
# Load the COVID-19 dataset into a DataFrame
covid_df = pd.read_csv(r"C:\Users\moizr\Downloads\covid.csv")  # Replace 'covid_dataset.csv' with the actual file path
covid_df.head(2)
```

```python
# Q. 1) Show the number of Confirmed, Deaths, and Recovered cases in each Region
cases_by_region = covid_df.groupby('Name of State / UT')[['Total Confirmed cases', 'Death', 'New recovered']].sum()
print("Number of Confirmed, Deaths, and Recovered cases in each Region:")
print(cases_by_region)
```

```python
# Q. 2) Remove all the records where the Confirmed Cases is Less Than 10
covid_df_filtered = covid_df[covid_df['Total Confirmed cases'] >= 10]
covid_df_filtered .head(2)
```

```python
# Q. 3) In which Region, the maximum number of Confirmed cases were recorded?
max_confirmed_region = covid_df.loc[covid_df['Total Confirmed cases'].idxmax()]['Name of State / UT']
```

```python
max_confirmed_region
```

```python
# Q. 4) In which Region, the minimum number of Deaths cases were recorded?
min_deaths_region = covid_df.loc[covid_df['Death'].idxmin()]['Name of State / UT']
```

```python
min_deaths_region
```

```python
# Q. 5) How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 2020?
india_cases = covid_df[covid_df['Name of State / UT'] == 'India'].sum()[['Total Confirmed cases', 'Death', 'New recovered']]
```

```python
india_cases
```

```python
# Q. 6-A) Sort the entire data wrt No. of Confirmed cases in ascending order
sorted_by_confirmed = covid_df.sort_values(by='Total Confirmed cases')
sorted_by_confirmed
```

```python
# Q. 6-B) Sort the entire data wrt No. of Recovered cases in descending order
sorted_by_recovered = covid_df.sort_values(by='New recovered', ascending=False)
```
